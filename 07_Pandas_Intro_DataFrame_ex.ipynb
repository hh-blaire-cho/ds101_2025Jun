---
Convert a gender column to have 'M' and 'F' instead of 'male' and 'female'
- use replace 


````python
# your code goes here
marvel_df["gender"] = marvel_df["gender"].replace({"male": "M", "female": "F"})
marvel_df
````

---
Let's roll it back. Convert a gender column to have 'male' and 'female'
- use lambda function


````python
marvel_df["gender"] = marvel_df["gender"].map(lambda x : "male" if x == "M" else "female")
marvel_df
````

---
Given the `marvel_df` pandas DataFrame, consider the conditions.
- first_appearance is before 1969(not including 1969)
- male
- team name not Fantastic Four

And find:
- find how many characters are qualified
- find their average of popularity score

````python
# your code goes here
c1 = (marvel_df["first_appearance"] < 1969)
c2 = (marvel_df["gender"] == "male")
c3 = (marvel_df["team"] != "Fantastic Four")

ans = marvel_df.loc[c1 & c2 & c3]
ans
````

````python
print(ans.shape[0])
print(ans["popularity_score"].mean().round(2))
````
